```julia
function DE!(dC,C,p,t)  #function to define process model
λ1,λ2,δ,KK=p #parameter vector for the process model
S=C[1]+C[2]
dC[1]=λ1*C[1]*(1.0-S/KK) #define differential equations
dC[2]=λ2*C[2]*(1.0-S/KK)-δ*C[2]*C[1]/KK; #define differential equations
end
```

```julia
function odesolver(t1,λ1,λ2,δ,KK,C01,C02) #function to solve the process model
p=(λ1,λ2,δ,KK) # parameter vector for the process model
C0=[C01,C02]
tspan=(0.0,maximum(t1)) #time horizon
prob=ODEProblem(DE!,C0,tspan,p) #define ODE model
sol=solve(prob,saveat=t1) #solve and save solutions
return sol[1,:],sol[2,:]
end
```

```julia
function odesolver2(t1,λ1,λ2,δ,KK,C01,C02) #function to solve the process model
p=(λ1,λ2,δ,KK) # parameter vector for the process model
C0=[C01,C02]
tspan=(0.0,maximum(t1)) #time horizon
prob=ODEProblem(DE!,C0,tspan,p) #define ODE model
sol=solve(prob) #solve and save solutions, including time
tt=sol.t[:]
return sol[1,:],sol[2,:],tt
end
```

```julia
function model(t1,a) #function to solve the process model
y1=zeros(length(t1))
y2=zeros(length(t1))
(y1,y2)=odesolver(t1,a[1],a[2],a[3],a[4],a[5],a[6]) #solve the process model
return y1,y2
end
```

```julia
function error(data11,data12,a) #function to evaluate the loglikelihood for the data given parameters a
y1=zeros(length(t1))
y2=zeros(length(t1))
(y1,y2)=model(t1,a);
e=0.0;
f=0.0;
dist1=Normal(0.0,a[7]);
e=loglikelihood(dist1,data11-y1)+loglikelihood(dist1,data12-y2);
return e
end
```

```julia
function fun(a) #function to evaluate the loglikelihood as a function of the parameters a 
return error(data11,data12,a)
end
```

```julia
function optimise(fun,θ₀,lb,ub;  #NLopt routine to maximise the function fun, with parameter estimates \theta_0 subject to bound constraints lb, ub
    dv = false,
    method = dv ? :LD_LBFGS : :LN_BOBYQA,
)

if dv || String(method)[2] == 'D'
    tomax = fun
else
    tomax = (θ,∂θ) -> fun(θ)
end

opt = Opt(method,length(θ₀))
opt.local_optimizer = Opt(:LN_NELDERMEAD, length(θ₀))
opt.max_objective = tomax
opt.lower_bounds = lb       # Lower bound
opt.upper_bounds = ub       # Upper bound
res = optimize(opt,θ₀)
return res[[2,1]]
end
```

```julia
θG =[λ1g,  λ2g,  δg,  KKg,   C0g[1], C0g[2], σg] #parameter estimates
lb=[0.00,  0.00, 0.00, 60.0,  0.0, 0.0, 0.0]; #lower bound
ub=[0.01,  0.01, 0.01, 90.0, 1.0, 1.0, 3.0]; #upper bound
(xopt,fopt)  = optimise(fun,θG,lb,ub) #call numerical optimization routine to give the vector of parameters xopt, and the maximum loglikelihood fopt
fmle=fopt
λ1mle=xopt[1]
λ2mle=xopt[2]
δmle=xopt[3]
Kmle=xopt[4]
C01mle=xopt[5]
C02mle=xopt[6]
σmle=xopt[7]
xopt
```

```julia
#Profile λ1
nptss=40 #points on the profile
λ1min=0.001 #lower bound for the profile - can be changed easily
λ1max=0.004 #upper bound for the profile - can be changed easily
λ1range=LinRange(λ1min,λ1max,nptss)  #vector of λ1 values along the profile
nrange=zeros(6,nptss); #matrix to store the nuisance parameters once optimized out
llλ1=zeros(nptss); #loglikelihood at each point along the profile
nllλ1=zeros(nptss); #normalised loglikelihood at each point along the profile
predictC1λ1=zeros(length(tt),nptss) #vector to store predictions at each point along the profile
predictC2λ1=zeros(length(tt),nptss) #vector to store predictions at each point along the profile


for i in 1:nptss
function fun1(aa) #function to return loglikelihood by fixing the interest parameter along the profile
return error(data11,data12,[λ1range[i],aa[1],aa[2],aa[3],aa[4],aa[5],aa[6]])
end

lb1=    [λ2mle,δmle,Kmle,C01mle,C02mle,σmle]/2.0; #lower bound
ub1=2.0*[λ2mle,δmle,Kmle,C01mle,C02mle,σmle]; #upper bound
θG1 = [λ2mle,δmle,Kmle,C01mle,C02mle,σmle];  #initial estimate - take the MLE 
local (xo,fo)=optimise(fun1,θG1,lb1,ub1) #numerical optimisation
nrange[:,i]=xo[:] #store the nuisance parameters
llλ1[i]=fo[1] #store the loglikelihood
(predictC1λ1[:,i],predictC2λ1[:,i])=model(tt,[λ1range[i],nrange[1,i],nrange[2,i],nrange[3,i],nrange[4,i],nrange[5,i],nrange[6,i]])
end

nllλ1=llλ1.-maximum(llλ1); #calculate normalised loglikelihood



C1λ1p=zeros(length(tt))
C1λ1m=1000*ones(length(tt))
C2λ1p=zeros(length(tt))
C2λ1m=1000*ones(length(tt))


for i in 1:nptss
if nllλ1[i] >= -1.921 #for all predictions where the normalised loglikelihood is above the threshold calculate the maximum/minimum predictions
    for j in 1:length(tt)
    C1λ1p[j]=max(predictC1λ1[j,i],C1λ1p[j])
    C1λ1m[j]=min(predictC1λ1[j,i],C1λ1m[j])
    C2λ1p[j]=max(predictC2λ1[j,i],C2λ1p[j])
    C2λ1m[j]=min(predictC2λ1[j,i],C2λ1m[j])

        end
end

end

s1=plot(λ1range,nllλ1,xlim=(λ1min,λ1max),ylim=(-3,0.1),lw=4,xlabel="λ1",ylabel="ll",legend=false)
s1=hline!([-1.92],lw=4)
s1=vline!([λ1mle],lw=4)
display(s1)
```
